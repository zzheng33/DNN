```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import seaborn as sns

res_dir="./diff_gpu_result/"
corun_dir="./corun_res/"
corun_shareCPU_dir="./corun_shareCPU_res/"
diff_cpu_dir = "./diff_cpu_result/"
diff_cpu_dir_npb="./NPB_result/"
npb_papi = "./NPB_result/papi/pure_result/"
cpu_affinity_DL ="./cpu_affinity_DL/"

def plot_diff_gpu(file_name):
    # Read the CSV file
    df = pd.read_csv(res_dir+file_name)
    
    # Get unique models and number of GPUs
    models = df["Model"].unique()
    num_gpus = df["Num_GPUs"].unique()

    # Set bar width and positions
    bar_width = 0.15
    bar_positions = np.arange(len(models))

    # Set the style and colors for the bars
    plt.style.use("ggplot")
    colors = plt.cm.viridis(np.linspace(0, 1, len(num_gpus)))

    # Plot the grouped bar chart
    plt.figure(figsize=(12, 6))
    for i, (gpu, color) in enumerate(zip(num_gpus, colors)):
        gpu_data = df[df["Num_GPUs"] == gpu]
        plt.bar(bar_positions + i * bar_width, gpu_data["Images/s"], width=bar_width, label=f"{gpu} GPUs", color=color)

    plt.xticks(bar_positions + (len(num_gpus) - 1) * bar_width / 2, models)
    plt.xlabel("Model")
    plt.ylabel("Images/s")

    # Extract batch size and number of workers from file name
    batch_size, num_workers = re.findall(r"\d+", file_name)
    plt.title(f"Training Throughput using multiple GPUs (Batch_size={batch_size}, # of workers={num_workers})")
    plt.legend()

    # Save the plot as an image file
    plt.savefig(f"./plot_images/multiGPU_{batch_size}_{num_workers}_.png", bbox_inches='tight')

    # Show the plot
    plt.show()
    

def plot_diff_batch(file_name):
    # Read the CSV file
    df = pd.read_csv(res_dir+file_name)
    
    # Get unique models and number of GPUs
    models = df["Model"].unique()
    num_gpus = df["Num_GPUs"].unique()
    batch_size = df["Batch_Size"].unique()

    # Set bar width and positions
    bar_width = 0.15
    bar_positions = np.arange(len(models))

    # Set the style and colors for the bars
    plt.style.use("ggplot")
    colors = plt.cm.viridis(np.linspace(0, 1, len(batch_size)))

    # Plot the grouped bar chart
    plt.figure(figsize=(12, 6))
    for i, (batch, color) in enumerate(zip(batch_size, colors)):
        batch_size_data = df[df["Batch_Size"] == batch]
        plt.bar(bar_positions + i * bar_width, batch_size_data["Images/s"], width=bar_width, label=f"{batch} Batch_Size", color=color)

    plt.xticks(bar_positions + (len(batch_size) - 1) * bar_width / 2, models)
    plt.xlabel("Model")
    plt.ylabel("Images/s")

    # Extract batch size and number of workers from file name
    gpu, num_workers = re.findall(r"\d+", file_name)
    plt.title(f"Training Throughput using various batch_size (# of GPU={gpu}, # of workers={num_workers})")
    plt.legend()

    # Save the plot as an image file
    plt.savefig(f"./plot_images/multiGPU_{gpu}_{num_workers}_.png", bbox_inches='tight')

    # Show the plot
    plt.show()
    



def plot_coRun(file_name, shareCPU):
    data = None
    if shareCPU == 0:
        data = pd.read_csv(corun_dir+file_name)
    else:
        data = pd.read_csv(corun_shareCPU_dir+file_name)

    n_bars_per_group = 2
    bar_width = 0.07

    # Set the style for the plot
    plt.style.use("ggplot")

    fig, ax = plt.subplots(figsize=(12, 6))

    def label_exists(label, handles_labels):
        _, labels = handles_labels
        return label in labels

    first_model = data.iloc[0]['Model']

    first_row = data.iloc[0]
    ax.bar(0, first_row['Images/s'], bar_width, color=plt.cm.viridis(0), label=first_model)

    for i in range(1, len(data), n_bars_per_group):
        group_data = [data.iloc[i], data.iloc[i + 1]]
        group_data.sort(key=lambda row: row['Model'] != first_model)

        for j, row in enumerate(group_data):
            x = (i // n_bars_per_group + 1) * (bar_width * n_bars_per_group + 0.1) + j * bar_width
            color = plt.cm.viridis(0) if row['Model'] == first_model else plt.cm.viridis((i + j) / len(data))
            bar = ax.bar(x, row['Images/s'], bar_width, color=color)
            if not label_exists(row['Model'], ax.get_legend_handles_labels()):
                bar.set_label(row['Model'])

    xticks = [bar_width / 2] + [(i + 1) * (bar_width * n_bars_per_group + 0.08) + bar_width for i in range(len(data) // n_bars_per_group)]
    ax.set_xticks(xticks)
    ax.set_xticklabels([first_model] + [f"{data.iloc[i]['Model']} / {data.iloc[i+1]['Model']}" for i in range(1, len(data), n_bars_per_group)])

    plt.xlabel('Groups', fontsize=12)
    plt.ylabel('Images/s', fontsize=12)

    plt.title(file_name[:-4])

    plt.xticks(rotation=45, fontsize=10)

    ax.legend(fontsize=10)

    ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()

    
def plot_diff_cpu(file_name):
    data = pd.read_csv(diff_cpu_dir+file_name)

    # Select required columns and sort by 'Model' and 'Num_Workers'
    data = data[['Model', 'Num_Workers', 'Images/s']].sort_values(by=['Model', 'Num_Workers'])

    # Get unique model names and the number of unique workers
    unique_models = data['Model'].unique()
    unique_workers = data['Num_Workers'].unique()

    # Set the style for the plot
    plt.style.use('ggplot')

    # Set up the line chart
    fig, ax = plt.subplots(figsize=(12, 6))

    # Plot the lines for each model
    for model in unique_models:
        model_data = data[data['Model'] == model]
        ax.plot(model_data['Num_Workers'], model_data['Images/s'], marker='o', label=model)

    # Set the x-axis labels
    ax.set_xticks(unique_workers)

    # Set axis labels and font size
    plt.xlabel('Num_Cores', fontsize=12)
    plt.ylabel('Images/s', fontsize=12)

    # Add a title and set the font size
    plt.title('Model Performance by Number of Cores')

    # Rotate the x-axis labels for better readability
    plt.xticks(fontsize=10)

    # Add a legend with unique items
    ax.legend(fontsize=10)

    # Add gridlines
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Display the chart
    plt.tight_layout()
    plt.show()
    
    
def plot_diff_cpu_NPB_OMP(file_name):
    data = pd.read_csv(diff_cpu_dir_npb+file_name)

    # Select required columns and sort by 'Model' and 'Num_Workers'
    data = data[['Application', 'Cores', 'Running Time']].sort_values(by=['Application', 'Cores'])

    # Get unique model names and the number of unique workers
    unique_models = data['Application'].unique()
    unique_workers = data['Cores'].unique()

    # Set the style for the plot
    plt.style.use('ggplot')

    # Set up the line chart
    fig, ax = plt.subplots(figsize=(12, 6))

    # Plot the lines for each model
    for model in unique_models:
        model_data = data[data['Application'] == model]
        ax.plot(model_data['Cores'], model_data['Running Time'], marker='o', label=model)

    # Set the x-axis labels
    ax.set_xticks(unique_workers)

    # Set axis labels and font size
    plt.xlabel('Num_Cores', fontsize=12)
    plt.ylabel('Ruuning Time(s)', fontsize=12)

    # Add a title and set the font size
    plt.title('App Performance by Number of Cores')

    # Rotate the x-axis labels for better readability
    plt.xticks(fontsize=10)

    # Add a legend with unique items
    ax.legend(fontsize=10)

    # Add gridlines
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Display the chart
    plt.tight_layout()
    plt.show()
    

def plot_npb_papi(file_name):
    data = pd.read_csv(npb_papi+file_name)
    
   # Calculate the averages for the first half and the second half of the rows
    first_half_avg = data.iloc[:len(data)//2].mean()
    second_half_avg = data.iloc[len(data)//2:].mean()

    # Set the style for the plot
    plt.style.use('ggplot')

    # Set up the subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Set the bar width
    bar_width = 0.2

    # Plot the first column in the first subplot
    ax1.bar(["Pack", "Spread"], [first_half_avg[0], second_half_avg[0]], width=bar_width, color=["steelblue", "indianred"], alpha=0.8)
    ax1.set_title(data.columns[0], fontsize=14)
    ax1.set_ylabel('Counts', fontsize=12)
    ax1.grid(axis='y', linestyle='--', alpha=0.7)

    # Plot the second column in the second subplot
    ax2.bar(["Pack", "Spread"], [first_half_avg[1], second_half_avg[1]], width=bar_width, color=["steelblue", "indianred"], alpha=0.8)
    ax2.set_title(data.columns[1], fontsize=14)
    ax2.set_ylabel('Counts', fontsize=12)
    ax2.grid(axis='y', linestyle='--', alpha=0.7)

    # Add a title for the entire figure
    fig.suptitle('Average Counts of Data Cache Miss', fontsize=16, y=1.05)

    # Display the chart
    plt.tight_layout()
    plt.show()
    
def plot_npb_papi_errorBox(file_name):
    data = pd.read_csv(npb_papi+file_name)
    
    # Extract the app and x from the file name
    match = re.search(r'papi_res_(\w+_\w+)', file_name)
    app_x = match.group(1) if match else "Unknown"
    
    first_half = data.iloc[:len(data)//2]
    second_half = data.iloc[len(data)//2:]

    # Set the style for the plot
    plt.style.use('ggplot')

    # Set up the subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Plot the first column in the first subplot as a box plot with thinner bars
    ax1.boxplot([first_half.iloc[:, 0], second_half.iloc[:, 0]], labels=["Pack", "Spread"], widths=0.3, patch_artist=True, boxprops=dict(facecolor="steelblue", alpha=0.8), medianprops=dict(color="black"))
    ax1.set_title(data.columns[0], fontsize=14)
    ax1.set_ylabel('Counts', fontsize=12)
    ax1.grid(axis='y', linestyle='--', alpha=0.7)

    # Plot the second column in the second subplot as a box plot with thinner bars
    ax2.boxplot([first_half.iloc[:, 1], second_half.iloc[:, 1]], labels=["Pack", "Spread"], widths=0.3, patch_artist=True, boxprops=dict(facecolor="indianred", alpha=0.8), medianprops=dict(color="black"))
    ax2.set_title(data.columns[1], fontsize=14)
    ax2.set_ylabel('Counts', fontsize=12)
    ax2.grid(axis='y', linestyle='--', alpha=0.7)

    # Add a title for the entire figure
    fig.suptitle(f'DCM for {app_x}', fontsize=16, y=1.05)

    # Display the chart
    plt.tight_layout()
    plt.show()
    
    
def plot_affinity_speedup(file_name):
    data = pd.read_csv(diff_cpu_dir_npb+file_name)
    data = data[['Application', 'Running Time', 'Spread']]

    # Set the style for the plot
    plt.style.use('ggplot')

    # Set up the bar chart
    bar_width = 0.1
    n_bars_per_group = 2
    n_groups = len(data) // n_bars_per_group
    fig, ax = plt.subplots(figsize=(12, 6))

    # Create a custom legend
    custom_legend = []

    for i in range(n_groups):
        # Calculate the ratio of the running times for the current pair of rows
        base = data.iloc[i * n_bars_per_group]['Running Time']
        ratio = data.iloc[i * n_bars_per_group + 1]['Running Time'] / base

        # Set the color based on the 'Spread' value
        color1 = "steelblue" if data.iloc[i * n_bars_per_group]['Spread'] == 0 else "indianred"
        color2 = "steelblue" if data.iloc[i * n_bars_per_group + 1]['Spread'] == 0 else "indianred"

        # Plot the bars for the current group
        x = i * (bar_width * n_bars_per_group + 0.1)
        bar1 = ax.bar(x, 1, bar_width, color=color1, alpha=0.8)
        bar2 = ax.bar(x + bar_width, ratio, bar_width, color=color2, alpha=0.8)

        # Add bars to the custom legend if their colors are not already in the legend
        if bar1[0].get_facecolor() not in [item[0].get_facecolor() for item in custom_legend]:
            custom_legend.append(bar1)
        if bar2[0].get_facecolor() not in [item[0].get_facecolor() for item in custom_legend]:
            custom_legend.append(bar2)

    # Set the x-axis labels
    ax.set_xticks([(bar_width * n_bars_per_group + 0.1) * i + bar_width / 2 for i in range(n_groups)])
    ax.set_xticklabels([f"{data.iloc[i]['Application']}" for i in range(0, len(data), n_bars_per_group)], rotation=45)

    # Set axis labels and font size
    plt.ylabel('Running Time Ratio', fontsize=12)

    # Add a title and set the font size
    plt.title('Running Time Ratios of Application')

    # Add the custom legend
    ax.legend(custom_legend, ['Pack', 'Spread'], fontsize=10)

    # Add gridlines
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Display the chart
    plt.tight_layout()
    plt.show()
    
    
def plot_cpu_affinity_DL(file_name):
    # Load the data
    data = pd.read_csv(cpu_affinity_DL+file_name)

    # Add a new column to the dataframe to distinguish between the first and second occurrences
    data['Label'] = data.groupby(['Model', 'Batch_Size', 'Num_Workers', 'Num_GPUs']).cumcount()
    data['Label'] = data['Label'].map({0: 'pack', 1: 'spread'})

    # Unique batch sizes
    batch_sizes = data['Batch_Size'].unique()

    # Create a figure and axes
    fig, axs = plt.subplots(len(batch_sizes), figsize=(10, len(batch_sizes)*5))

    palette = {'pack': '#206C5A', 'spread': '#219176'}

    # For each batch size, create a bar plot
    for i, batch_size in enumerate(batch_sizes):
        subset = data[data['Batch_Size'] == batch_size]
        sns.barplot(x='Model', y='Images/s', hue='Label', data=subset, palette=palette, ax=axs[i])
        axs[i].set_title(f'Batch Size: {batch_size}')
        axs[i].grid(axis='y', linestyle='--')


    # Adjust the layout
    plt.tight_layout()
    plt.show()
```
